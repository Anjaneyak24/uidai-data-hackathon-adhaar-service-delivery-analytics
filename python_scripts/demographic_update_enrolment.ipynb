```python
print("\n=== AADHAAR DEMOGRAPHIC ENROLMENT KEY INSIGHTS SUMMARY ===\n")

# ------------------- Overall Metrics -------------------
total_records = len(df)

total_demo_enrol = df['total_demo_enrol'].sum()
total_children = df['age 5-17'].sum()
total_adults = df['above 17'].sum()

avg_monthly_demo_enrol = (
    df.groupby('month_name')['total_demo_enrol']
      .sum()
      .mean()
)

print("OVERALL DEMOGRAPHIC PERFORMANCE:\n")
print(f"Total Records: {total_records}")
print(f"Total Demographic Enrolments: {total_demo_enrol}")
print(f"Children (5–17 years): {total_children}")
print(f"Adults (18+ years): {total_adults}")
print(f"Average Monthly Demographic Enrolment: {avg_monthly_demo_enrol:.0f}")

# ------------------- Top States -------------------
top_states = (
    df.groupby('state_clean')['total_demo_enrol']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("\nTOP 5 STATES BY DEMOGRAPHIC ENROLMENT:\n")
for i, (state, enrol) in enumerate(top_states.items(), 1):
    print(f"{i}. {state}: {enrol}")

# ------------------- Bottom States -------------------
bottom_states = (
    df.groupby('state_clean')['total_demo_enrol']
      .sum()
      .sort_values(ascending=True)
      .head(5)
)

print("\nBOTTOM 5 STATES BY DEMOGRAPHIC ENROLMENT:\n")
for i, (state, enrol) in enumerate(bottom_states.items(), 1):
    print(f"{i}. {state}: {enrol}")

# ------------------- Top Districts -------------------
top_districts = (
    df.groupby('district_clean')['total_demo_enrol']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("\nTOP 5 DISTRICTS BY DEMOGRAPHIC ENROLMENT:\n")
for i, (district, enrol) in enumerate(top_districts.items(), 1):
    print(f"{i}. {district}: {enrol}")

# ------------------- Age Group Distribution -------------------
child_percent = 100 * total_children / total_demo_enrol
adult_percent = 100 * total_adults / total_demo_enrol

print("\nAGE GROUP DISTRIBUTION (DEMOGRAPHIC):\n")
print(f"Children (5–17 years): {total_children} ({child_percent:.1f}%)")
print(f"Adults (18+ years): {total_adults} ({adult_percent:.1f}%)")

# ------------------- Monthly Trend -------------------
monthly_demo = df.groupby('month_name')['total_demo_enrol'].sum()

max_month = monthly_demo.idxmax()
min_month = monthly_demo.idxmin()

print("\nMONTHLY DEMOGRAPHIC TREND:\n")
print(f"Month with Highest Demographic Enrolment: {max_month} ({monthly_demo[max_month]})")
print(f"Month with Lowest Demographic Enrolment: {min_month} ({monthly_demo[min_month]})")
```
