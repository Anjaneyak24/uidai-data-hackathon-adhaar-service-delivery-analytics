```python
import pandas as pd
import matplotlib.pyplot as plt

age_cols = ['age 5-17', 'above 17']

df_state_age = df.groupby('state_clean')[age_cols].sum().reset_index()

df_state_age['total_enrol'] = df_state_age[age_cols].sum(axis=1)

df_state_age = df_state_age.sort_values('total_enrol', ascending=False)

# Top 20 → Millions
top_20 = df_state_age.head(20).copy()
top_20[age_cols] = top_20[age_cols] / 1_000_000

# Bottom 16 → Thousands
bottom_16 = df_state_age.tail(16).copy()
bottom_16[age_cols] = bottom_16[age_cols] / 1_000

def plot_state_biometric(df_plot, title, unit='M'):
    plt.figure(figsize=(14,7))

    plt.bar(df_plot['state_clean'], df_plot['age 5-17'],
            label='Age 5–17', color='#FE3B00')

    plt.bar(df_plot['state_clean'], df_plot['above 17'],
            bottom=df_plot['age 5-17'],
            label='Above 17', color='#0E169E')

    ylabel = 'Total Biometric Enrolment (Millions)' if unit == 'M' else 'Total Biometric Enrolment (Thousands)'
    plt.ylabel(ylabel, fontsize=12)

    plt.title(title, fontsize=16)
    plt.xlabel('State', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Age Group')

    # ---- Data labels ----
    for i, row in enumerate(df_plot.itertuples()):
        if unit == 'M':
            plt.text(i, row._2 / 2, f"{row._2:.2f} M",
                     ha='center', va='center', color='white', fontsize=9)
            plt.text(i, row._2 + row._3 / 2, f"{row._3:.2f} M",
                     ha='center', va='center', color='white', fontsize=9)
        else:
            plt.text(i, row._2 / 2, f"{row._2:.1f}K",
                     ha='center', va='center', color='white', fontsize=9)
            plt.text(i, row._2 + row._3 / 2, f"{row._3:.1f}K",
                     ha='center', va='center', color='white', fontsize=9)

    plt.tight_layout()
    plt.show()
    
plot_state_biometric(
    top_20,
    'Top 20 States: Biometric Enrolment by Age Group (in Millions)',
    unit='M'
)
plot_state_biometric(
    bottom_16,
    'Bottom 16 States: Biometric Enrolment by Age Group (in Thousands)',
    unit='K'
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ---- Aggregate biometric enrolment by district ----
age_cols = ['age 5-17', 'above 17']
df_district_age = df.groupby('district_clean')[age_cols].sum().reset_index()

# Total enrolment per district
df_district_age['total_enrol'] = df_district_age[age_cols].sum(axis=1)

# Sort descending by total enrolment
df_district_age = df_district_age.sort_values('total_enrol', ascending=False)

# Top 30 districts → Thousands
top_30 = df_district_age.head(30).copy()
top_30[age_cols] = top_30[age_cols] / 1_000  # convert to Thousands

# Bottom 30 districts → Absolute numbers
bottom_30 = df_district_age.tail(30).copy()

# ---- Function to plot stacked bar chart with labels ----
def plot_stacked_bar(df_plot, title, millions=True):
    plt.figure(figsize=(16,6))
    
    # Stacked bars
    bars1 = plt.bar(df_plot['district_clean'], df_plot['age 5-17'], label='Age 5-17', color='#6B007B')
    bars2 = plt.bar(df_plot['district_clean'], df_plot['above 17'], bottom=df_plot['age 5-17'], label='Above 17', color='#12239E')

    # Titles & labels
    plt.title(title, fontsize=16)
    plt.xlabel('District', fontsize=12)
    ylabel = 'Total Biometric Enrolment (Thousands)' if millions else 'Total Biometric Enrolment'
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Age Group')

    # Add data labels inside bars
    for i, row in enumerate(df_plot.itertuples()):
        if millions:
            # Integer + K for thousands
            plt.text(i, row._2/2, f"{int(row._2)}K", ha='center', va='center', color='white', fontsize=9)
            plt.text(i, row._2 + row._3/2, f"{int(row._3)}K", ha='center', va='center', color='white', fontsize=9)
        else:
            # Absolute numbers with commas
            plt.text(i, row._2/2, f"{int(row._2):,}", ha='center', va='center', color='white', fontsize=9)
            plt.text(i, row._2 + row._3/2, f"{int(row._3):,}", ha='center', va='center', color='white', fontsize=9)

    # Adjust layout
    plt.tight_layout()
    plt.show()

# ---- Plot Top 30 Districts (in Thousands) ----
plot_stacked_bar(top_30, 'Top 30 District Biometric Enrolment by Age Group (Thousands)', millions=True)

# ---- Plot Bottom 30 Districts (Absolute Numbers) ----
plot_stacked_bar(bottom_30, 'Bottom 30 District Biometric Enrolment by Age Group', millions=False)
```

```python
## Extremely low enrolment pincodes represent non-residential or special-use
## administrative zones rather than data errors, and are therefore
## analytically meaningful outliers.

# -----------------------------
# Define age columns (actual column names)
# -----------------------------
age_cols = ['age 5-17', 'above 17']

# -----------------------------
# Aggregate data at pincode level
# -----------------------------
df_pincode = (
    df.groupby('pincode')[age_cols]
      .sum()
      .reset_index()
)

# Ensure pincode is treated as identifier
df_pincode['pincode'] = df_pincode['pincode'].astype(str)

# -----------------------------
# Total biometric enrolment
# -----------------------------
df_pincode['total_enrol'] = df_pincode[age_cols].sum(axis=1)

# -----------------------------
# Data-driven low-enrolment threshold (5th percentile)
# -----------------------------
low_threshold = df_pincode['total_enrol'].quantile(0.05)

# -----------------------------
# Summary statistics
# -----------------------------
total_pincodes = df_pincode['pincode'].nunique()
low_total_count = (df_pincode['total_enrol'] < low_threshold).sum()

age_5_17_low = (df_pincode['age 5-17'] < low_threshold).sum()
age_18_plus_low = (df_pincode['above 17'] < low_threshold).sum()

age_5_17_0_1 = df_pincode['age 5-17'].isin([0, 1]).sum()
age_18_plus_0_1 = df_pincode['above 17'].isin([0, 1]).sum()

# -----------------------------
# Bottom 30 pincodes (critical low enrolment)
# -----------------------------
bottom_30 = (
    df_pincode
    .sort_values('total_enrol')
    .head(30)[['pincode'] + age_cols + ['total_enrol']]
)

# -----------------------------
# Output
# -----------------------------
print("SUMMARY INSIGHTS — TOTAL BIOMETRIC ENROLMENT")
print("-" * 60)
print(f"Total pincodes analysed              : {total_pincodes}")
print(f"5th percentile threshold             : {low_threshold:.0f}")
print(f"Pincodes below threshold             : {low_total_count}")
print(f"Age 5–17 below threshold             : {age_5_17_low}")
print(f"Age 18+ below threshold              : {age_18_plus_low}")
print(f"Age 5–17 values = 0 or 1              : {age_5_17_0_1}")
print(f"Age 18+ values = 0 or 1               : {age_18_plus_0_1}")

print("\nBottom 30 Pincodes Biometric Enrolment")
print("-" * 60)
print(bottom_30)
```

```python
print("\n=== AADHAAR BIOMETRIC ENROLMENT KEY INSIGHTS SUMMARY ===\n")

# ------------------- Overall Metrics -------------------
total_records = len(df)

total_biometric_enrol = df['total_biometric_enrol'].sum()
total_children = df['age 5-17'].sum()
total_adults = df['above 17'].sum()

avg_monthly_biometric_enrol = (
    df.groupby('month_name')['total_biometric_enrol']
      .sum()
      .mean()
)

print("OVERALL BIOMETRIC PERFORMANCE:\n")
print(f"Total Records: {total_records}")
print(f"Total Biometric Enrolments: {total_biometric_enrol}")
print(f"Children (5–17 years): {total_children}")
print(f"Adults (18+ years): {total_adults}")
print(f"Average Monthly Biometric Enrolment: {avg_monthly_biometric_enrol:.0f}")

# ------------------- Top States -------------------
top_states = (
    df.groupby('state_clean')['total_biometric_enrol']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("\nTOP 5 STATES BY BIOMETRIC ENROLMENT:\n")
for i, (state, enrol) in enumerate(top_states.items(), 1):
    print(f"{i}. {state}: {enrol}")

# ------------------- Bottom States -------------------
bottom_states = (
    df.groupby('state_clean')['total_biometric_enrol']
      .sum()
      .sort_values(ascending=True)
      .head(5)
)

print("\nBOTTOM 5 STATES BY BIOMETRIC ENROLMENT:\n")
for i, (state, enrol) in enumerate(bottom_states.items(), 1):
    print(f"{i}. {state}: {enrol}")

# ------------------- Top Districts -------------------
top_districts = (
    df.groupby('district_clean')['total_biometric_enrol']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("\nTOP 5 DISTRICTS BY BIOMETRIC ENROLMENT:\n")
for i, (district, enrol) in enumerate(top_districts.items(), 1):
    print(f"{i}. {district}: {enrol}")

# ------------------- Age Group Distribution -------------------
child_percent = 100 * total_children / total_biometric_enrol
adult_percent = 100 * total_adults / total_biometric_enrol

print("\nAGE GROUP DISTRIBUTION (BIOMETRIC):\n")
print(f"Children (5–17 years): {total_children} ({child_percent:.1f}%)")
print(f"Adults (18+ years): {total_adults} ({adult_percent:.1f}%)")

# ------------------- Monthly Trend -------------------
monthly_biometric = df.groupby('month_name')['total_biometric_enrol'].sum()

max_month = monthly_biometric.idxmax()
min_month = monthly_biometric.idxmin()

print("\nMONTHLY BIOMETRIC TREND:\n")
print(f"Month with Highest Biometric Enrolment: {max_month} ({monthly_biometric[max_month]})")
print(f"Month with Lowest Biometric Enrolment: {min_month} ({monthly_biometric[min_month]})")
```
